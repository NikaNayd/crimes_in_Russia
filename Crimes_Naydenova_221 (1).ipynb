## Data Clean-up

Before analysing the dataset it is necessary to asure that all the values are correct and there are not any NaNs or (in case there are) it is vital to remove such rows in order not to reduce the accurance of the analyse.

```python
df.isnull().sum()
```

As we can see from the table above, there are no empty cells. Let us check it one more time in a different way (just to be absolutely sure)

```python
k=0
s=0
with open(r'C:\Users\Nika\OneDrive\Desktop\crime.csv') as file:
    for line in file:
        a=line.split()
        for i in range (len(a)):
            s+=1
            if a[i]=='N':
                print('There exist Nan in the dataset')
                break
            else:
                k+=1

if k==s:
    print('There are no NaNs in the dataset')
```

Well, Dataset is already clean.

## Data transformation

Since we have cleaned up the Dataset we can transform it to simplify our following researsh.

```python
df.month=df.month.apply(lambda x: str(x)[-4:]).astype(int)
```

This is the way we rewrote years instead of the full dates. It was made to make next analyse easier

```python
c=[]
for i in range ( 205):
    if int(df['month'].loc[df.index[i]])==2008 or int(df['month'].loc[df.index[i]])==2014:
        c.append('Crises')
    else:
        c.append('Not_crises')
df.insert(1, 'Crises_or_not', c, allow_duplicates = True)
df.head(13)
```

## Observation

```python
import seaborn as sns
df.corr()
sns.heatmap(df.corr())
```

This heatmap shows how do different types of crimes depend on each other. If the cell is light, then two crimes are proportional, if dark - then improportional. In this heatmap I noticed 2 interesting facts:
1. How does Total number of crimes depend on the number of Thefts.
2. How does the number of Serious crimes depend on weapons.
In the next two plots I am going to analyse them more precisly

```python
import matplotlib.pyplot  as plt
plt.figure(figsize=(10, 10))
sns.lineplot(data=df[['Total_crimes', 'Theft', 'Huge_damage']])
plt.xticks(rotation=90)
plt.tight_layout()
plt.title('Correlation')
plt.ylabel('Number')
```

From the graph we can see that total crimes and thefts are proportional (when first rises or falls, the second one rises or falls as well). While huge damage is improportional with thefts and total crimes. For example, if there are many huge damages in month then comperatively there are not so many thefts and the total number of crimes.

```python
plt.figure(figsize=(20, 8))
df.groupby(by='month').max()
sns.lineplot(data=df.groupby(by='month').max(),sizes=(200, 200))
plt.title('Correlation of all types of crimes')
plt.ylabel('Number')
plt.xlabel('Year')
```

From this graph we can observe all types of crimes. We can see that the most frequent crimes are murders, serious crimes (which includes several types), huge damages, drugs and fraus scam.

```python
plt.figure(figsize=(17, 8))
plt.bar(df['month'], df['Weapons'], label='weapons')
plt.plot(df['month'], df['Hooligan'], label='Serious', c='r')
plt.xticks(rotation=90)
plt.legend()
plt.title('Correlatiom of Serious crimes and Weapons')
plt.ylabel('Number')
plt.xlabel('Year')
```
